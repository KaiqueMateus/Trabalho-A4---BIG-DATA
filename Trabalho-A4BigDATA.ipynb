```python
import pandas as pd
import matplotlib.pyplot as plt  

tabela = pd.read_excel(path)


tabela = tabela.drop(['Width', 'Length'], axis=1)

tabela_ordenada = tabela.sort_values(by='Sales_in_thousands', ascending=False)

top_5_mais_vendidos = tabela_ordenada.head(5)

top_5_mais_vendidos.loc[:, 'Ganhos'] = top_5_mais_vendidos['Sales_in_thousands'] * top_5_mais_vendidos['Price_in_thousands']

montadora_mais_lucrativa = top_5_mais_vendidos.groupby('Manufacturer')['Ganhos'].sum().idxmax()

total_lucro_montadora_mais_lucrativa = top_5_mais_vendidos.groupby('Manufacturer')['Ganhos'].sum().max()

print("Os cincos veículos mais vendidos são:")
display(top_5_mais_vendidos[['Manufacturer', 'Model', 'Sales_in_thousands', 'Price_in_thousands']])

print("Lucro/Ganhos em dinheiro de cada um dos cinco veículos mais vendidos:")
display(top_5_mais_vendidos[['Manufacturer', 'Model', 'Ganhos']])

print("Montadora mais lucrativa:", montadora_mais_lucrativa)

plt.bar(top_5_mais_vendidos['Manufacturer'], top_5_mais_vendidos['Ganhos'])
plt.xlabel('Manufacturer')
plt.ylabel('Ganhos')
plt.title('Marcas que mais lucraram com carros vendidos')
plt.show()

mean = round(top_5_mais_vendidos['Ganhos'].mean(), 2)
variance = round(top_5_mais_vendidos['Ganhos'].var(), 2)
std_deviation = round(top_5_mais_vendidos['Ganhos'].std(), 2)

estatisticas_df = pd.DataFrame({'Estatística': ['Média', 'Variância', 'Desvio Padrão'],
                                'Valor': [mean, variance, std_deviation]})

display(estatisticas_df)


```
